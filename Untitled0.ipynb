```python
import nest_asyncio
from telegram import Update, InlineKeyboardButton, InlineKeyboardMarkup
from telegram.ext import Application, CommandHandler, MessageHandler, filters, CallbackContext, CallbackQueryHandler
import asyncio
import re
from datetime import datetime, timedelta

# Apply nest_asyncio to fix event loop issue
nest_asyncio.apply()

# Your bot token (replace with your actual bot token)
TOKEN = "7675486045:AAGusZXRVOXynndK0isrjdOIRadxCK_2Icw"  # Replace with your token

# In-memory user data (for simplicity; replace with database in production)
user_data = {}

# Function to validate Ethereum address (MetaMask wallet address)
def is_valid_wallet_address(wallet_address: str) -> bool:
    # Check if the address starts with '0x' and is followed by exactly 40 hexadecimal characters
    return bool(re.match(r"^0x[a-fA-F0-9]{40}$", wallet_address))

# Start command handler
async def start(update: Update, context: CallbackContext):
    user_id = update.message.from_user.id
    # Initialize user data if not exists
    if user_id not in user_data:
        user_data[user_id] = {"balance": 0.0, "mining_progress": 0, "mining_count": 0, "wallet": None}

    # Get the mining count and wallet address for the user
    mining_count = user_data[user_id]["mining_count"]
    wallet_address = user_data[user_id]["wallet"]

    # Define inline keyboard
    keyboard = [
        [InlineKeyboardButton("Investment", callback_data="investment"),
         InlineKeyboardButton("Start Mining", callback_data="start_mining")],
        [InlineKeyboardButton("Coin Info", callback_data="coin_info"),
         InlineKeyboardButton("Set Wallet", callback_data="set_wallet")],
        [InlineKeyboardButton("Stats", callback_data="stats")]
    ]
    reply_markup = InlineKeyboardMarkup(keyboard)

    # Send a welcome message with the image and mining count
    welcome_message = f"به ربات قطرکوین خوش آمدید! شما {mining_count} کوین استخراج کرده‌اید. لطفاً یکی از گزینه‌ها را انتخاب کنید:"
    if wallet_address:
        welcome_message += f"\nآدرس کیف پول شما: {wallet_address}"
    else:
        welcome_message += "\nشما هنوز آدرس کیف پول خود را تنظیم نکرده‌اید. لطفاً در بخش 'کیف پول' آدرس خود را وارد کنید."

    photo_url = "https://i.imgur.com/mzeRO6g.jpg"  # Link to your image

    # Send the image and the welcome message
    await update.message.reply_photo(photo=photo_url, caption=welcome_message, reply_markup=reply_markup)

# Handlers for each option when inline button is pressed
async def button(update: Update, context: CallbackContext):
    query = update.callback_query
    user_id = query.from_user.id
    await query.answer()  # Answer the callback query to prevent it from timing out

    if query.data == "investment":
        investment_message = """
        💼 **فرصت‌های سرمایه‌گذاری در قطرکوین**

        قطرکوین به عنوان یک ارز دیجیتال در حال رشد با تمرکز بر منطقه شورای همکاری خلیج فارس (GCC) فرصت‌های عالی سرمایه‌گذاری را ارائه می‌دهد.
        """
        await query.message.reply_text(investment_message)
    elif query.data == "start_mining":
        await simulate_mining(update, context, user_id)
    elif query.data == "coin_info":
        coin_info = """
        **قطرکوین (QatarCoin)** یک ارز دیجیتال طراحی‌شده برای تسهیل تراکنش‌های مالی در منطقه شورای همکاری خلیج فارس (GCC) است.
        """
        await query.message.reply_text(coin_info)
    elif query.data == "set_wallet":
        await ask_for_wallet(update, context, user_id)
    elif query.data == "stats":
        await show_stats(update, context)  # Allow all users to see stats

# Show bot statistics: number of users and total mined coins
async def show_stats(update: Update, context: CallbackContext):
    # Count the number of users
    total_users = len(user_data)

    # Calculate the total number of coins mined by all users
    total_mined_coins = sum(user["mining_count"] for user in user_data.values())

    stats_message = f"""
    📊 **آمار ربات:**
    - تعداد کل اعضا: {total_users}
    - تعداد کل کوین‌های استخراج‌شده: {total_mined_coins}
    """

    await update.callback_query.message.reply_text(stats_message)

# Ask user to enter their wallet address
async def ask_for_wallet(update: Update, context: CallbackContext, user_id: int):
    # Ask the user for their wallet address
    await update.callback_query.message.reply_text("لطفاً آدرس کیف پول خود را وارد کنید (مثلاً آدرس اتریوم).")

# Save wallet address after validation
async def save_wallet_address(update: Update, context: CallbackContext, user_id: int):
    wallet_address = update.message.text
    # Validate the wallet address
    if is_valid_wallet_address(wallet_address):
        # Save wallet address if valid
        user_data[user_id]["wallet"] = wallet_address
        await update.message.reply_text(f"آدرس کیف پول شما با موفقیت تنظیم شد: {wallet_address}")
    else:
        # Send error message if address is invalid
        await update.message.reply_text("آدرس کیف پول وارد شده نادرست است. لطفاً آدرس صحیح متا ماسک خود را وارد کنید که با '0x' شروع شده و 40 کاراکتر داشته باشد.")

# Simulate mining process
async def simulate_mining(update: Update, context: CallbackContext, user_id: int):
    # Check if 12 hours have passed since the last mining
    last_mining_time = user_data.get(user_id, {}).get("last_mining_time", None)
    if last_mining_time and datetime.now() - last_mining_time < timedelta(hours=12):
        await update.callback_query.message.reply_text("شما فقط می‌توانید هر 12 ساعت یک بار استخراج کنید. لطفاً بعداً امتحان کنید.")
        return

    # Simulate mining process
    user_data[user_id]["mining_progress"] = 100  # Simulate a successful mining
    user_data[user_id]["balance"] += 1.0  # Add 1 coin to balance
    user_data[user_id]["mining_count"] += 1  # Increment mining count
    user_data[user_id]["last_mining_time"] = datetime.now()  # Update last mining time

    # Send mining completion message
    await update.callback_query.message.reply_text("استخراج با موفقیت انجام شد! 1 کوین به موجودی شما اضافه شد.")

# Set up the Telegram bot application
async def main():
    application = Application.builder().token(TOKEN).build()

    # Register handlers
    application.add_handler(CommandHandler("start", start))
    application.add_handler(CallbackQueryHandler(button))
    application.add_handler(MessageHandler(filters.TEXT & ~filters.COMMAND, save_wallet_address))

    # Run the bot
    await application.run_polling()

# Run the bot
asyncio.run(main())
```
